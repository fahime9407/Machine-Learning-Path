# Simple Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

### define dataset

```python
x = np.linspace(0, 50, 51).reshape(-1, 1)
y = np.linspace(0, 50, 51) + 5

x.shape, y.shape
```

### preprocessing

```python
plt.scatter(x, y, color="purple")
plt.title("Linear Regression Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
msk = np.random.rand(len(x)) < 0.8

x_train, x_test = x[msk], x[~msk]
y_train, y_test = y[msk], y[~msk]

x_train.shape, x_test.shape
```

### define model

```python
mymodel = LinearRegression()
mymodel.fit(x_train, y_train)

print(f"coefficient : {mymodel.coef_[0]:.6f} | intercept : {mymodel.intercept_:.6f}")
```

```python
plt.scatter(x, y, color="orange")
plt.plot(x, mymodel.intercept_ + mymodel.coef_[0] * x, color="green")
plt.title("Simple Linear Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
mymodel.predict([[25]])
```

```python
mymodel.predict([[11], [1.25], [35]])
```

### model evaluation

```python
y_pred = mymodel.predict(x_test)
```

```python
mse = np.mean((y_test - y_pred) ** 2)
mae = np.mean(np.absolute(y_test - y_pred))
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error : {mse} | Mean Absolute Error : {mae} | R2 Score : {r2}")
```
