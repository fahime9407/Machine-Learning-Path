## Drawing various linear and nonlinear charts.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Linear

```python
x = np.arange(-5.0, 5.0, 0.1)

y = 2 *(x) + 3
noise = 2 * np.random.normal(size=x.size)
ydata = y + noise

plt.plot(x, ydata, "o")
plt.plot(x, y, "r")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.show()
```

## Non Linear

### polynomial

```python
x = np.arange(-5.0, +5.0, 0.1)
y = 1 * (x ** 3) + 1 * (x ** 2) + 1 * x + 3
noise = 20 * np.random.normal(size=x.size)
ydata = y + noise

plt.plot(x, ydata, "o")
plt.plot(x, y, "r")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.show()
```

### Quadratic(polynomial)

```python
x = np.arange(-5.0, +5.0, +0.1)
y = np.power(x, 2)
noise = 2 * np.random.normal(size=x.size)
ydata = y + noise

plt.plot(x, ydata, "o")
plt.plot(x, y, "r")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.show()
```

### Exponential

```python
x = np.arange(-5.0, +5.0, +0.1)
y = np.exp(x)
noise = 10 * np.random.normal(size=x.size)
ydata = y + noise

plt.plot(x, ydata, "o")
plt.plot(x, y)
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.show()
```

### Logarithmic

```python
x = np.arange(0.1, 10.0, +0.1)
y = 5 * np.log(x)
noise = np.random.normal(size=x.size)
ydata = y + noise

plt.plot(x, ydata, "o")
plt.plot(x, y)
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.show()
```

### Sigmoidal or Logistic

```python
x = np.arange(-5.0, 5.0, 0.1)
y = 1 - 4 / (1 + np.power(3, x - 2))
noise = np.random.normal(size=x.size) / 3
ydata = y + noise

plt.scatter(x, ydata, color="green")
plt.plot(x, y)
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.show()
```
